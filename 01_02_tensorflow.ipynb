# manipulating tensors
**Basic operation**

```python
# + , - , * , /
import time as time
import random as random

tns3=tf.constant([[10,7],
                  [3,4]])
print(tns3 + 10)
print(tns3 - 5)
print(tns3 * 2)
print(tns3 / 3)

tns4 = tf.multiply(tns3, random.randint(1, 100)) + random.randint(1, 100)
```

## matrix multiplication
* most common operation in Tensors manipulation 

```python
tns1 * tns3
```

```python
tf.matmul(tns1,tns3)
```

```python
# matrix multiplication with @ :: python opariter
tns1 @ tns3
```

```python
# if the tensor are not combin we use reshalpe to make multplication avalible
multi_tns1 =tf.constant([[1,2],
                         [3,4],
                         [5,6]])
multi_tns2 =tf.constant([[7,8],
                         [9,10],
                         [11,12]])
multi_tns1 @ tf.reshape(multi_tns2,[2,3])
```

```python
# do it with transpose
# transpose make the line as colume and the colume as

multi_tns1 @ tf.transpose(multi_tns2)
```

```python
tf.tensordot(multi_tns1,tf.transpose(multi_tns2), axes=1,)

# to make multiplication with tensor :
#    1- with python parametr @ 
#    2- with tf.transpose
#    2- with tf.matmul
# to make reshape :
#    1- reshape()
#    2- transpose
#    3- 
```
